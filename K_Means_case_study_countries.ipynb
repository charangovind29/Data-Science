```python
data_new = df.drop(columns='country',axis=1)
```

```python
data_new
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
df_array = scaler.fit_transform(data_new)
data_scaled = pd.DataFrame(df_array,columns=data_new.columns)
```

```python
data_scaled
```

```python
data_scaled.describe().T
```

```python
from sklearn.cluster import KMeans
```

```python
sse = {}

for k in range(1,10):
    kmeans = KMeans(n_clusters=k, random_state=1).fit(data_scaled)
    
    sse[k] = kmeans.inertia_
```

```python
sse
```

```python
import warnings 
warnings.filterwarnings('ignore')
```

```python
plt.plot(list(sse.keys()),list(sse.values()),'yx-')
plt.xlabel('number of clusters')
plt.ylabel('sum of squared error')
```

```python
from sklearn.metrics import silhouette_score
```

```python
sc = {}

for k in range(2,10):
    kmeans = KMeans(n_clusters=k,random_state=1).fit(data_scaled)
    
    labels = kmeans.predict(data_scaled)
    
    sc[k] = silhouette_score(data_scaled,labels)
```

```python
sc
```

```python
plt.plot(list(sc.keys()),list(sc.values()),'ro-')
plt.xlabel('number of clusters')
plt.ylabel('silhouette score')
plt.show()
```

```python
kmeans = KMeans(n_clusters= 4,random_state= 1)

kmeans.fit(data_scaled)

data_scaled_copy = data_scaled.copy(deep=True)

data_scaled_copy['kmeans_label'] = kmeans.predict(data_scaled)

df['kmeans_label'] = kmeans.predict(data_scaled)
```
