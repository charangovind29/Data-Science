```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
df_scaled = sc.fit_transform(df_new)
```

```python
df_new.columns
```

```python
df_df1 = pd.DataFrame(df_scaled,columns=df_new.columns)
```

```python
df_df1
```

```python
from sklearn.cluster import KMeans
```

```python
clusters = list(range(2,8))
ssd = []

for num_cluster in clusters:
    model_clus = KMeans(n_clusters=num_cluster,max_iter=150,random_state=1)
    model_clus.fit(df_df1)
    ssd.append(model_clus.inertia_)
```

```python
ssd
```

```python
plt.plot(clusters,ssd)
```

```python
kmodel = KMeans(n_clusters=4,max_iter=150,random_state=1)
kmodel.fit(df_df1)
```

```python
kmodel.labels_
```

```python
df['cluster_id'] =kmodel.labels_
```
